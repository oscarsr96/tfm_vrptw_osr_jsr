```python
import pandas as pd
import random
import math

#Problem params

N_PROBLEMS = 10

N_CLIENTS = 20

MIN_CLIENT_TIME_WINDOW = 10
MAX_CLIENT_TIME_WINDOW = 50
MIN_Y = 0
MAX_Y = 100
MIN_X = 0
MAX_X = 100
MIN_CLEINT_DEMAND = 1
MAX_CLIENT_DEMAND = 10
MIN_CLIENT_START_TIME = 1
MAX_CLIENT_START_TIME = 50 * N_CLIENTS/10
MIN_CLIENT_END_TIME = 50
MAX_CLIENT_END_TIME = 100 * N_CLIENTS/10

COLUMNS = ['CUSTOMER', 'X', 'Y', 'DEMAND',
           'READY_TIME', 'DUE_TIME', 'SERVICE_TIME']
#Problem generation

for i in range(N_PROBLEMS):
  result = []
  for j in range(N_CLIENTS+1):
    if j == 0:
      result.append((0, 0,0, 0, 0, 0, 0))
    else:

      customer = j

      x = random.randint(MIN_X, MAX_X)
      y = random.randint(MIN_X, MAX_X)
      distance = math.sqrt(x*x+y*y)

      demand = random.randint(MIN_CLEINT_DEMAND, MAX_CLIENT_DEMAND)

      ready_time = random.randint(MIN_CLIENT_START_TIME, MAX_CLIENT_START_TIME)

      due_time = random.randint(max(ready_time, int(distance)),
                                MAX_CLIENT_END_TIME + int(distance))
      service_time = random.randint(MIN_CLIENT_TIME_WINDOW,
                                    MAX_CLIENT_TIME_WINDOW)

      result.append((customer, x, y, demand, ready_time, due_time, service_time))

  df = pd.DataFrame(result)
  df.columns = COLUMNS
  print(df)
  df.to_csv(f'./vrptw_{N_CLIENTS}_customers_problem_{i+1}.csv', index=False)




```
